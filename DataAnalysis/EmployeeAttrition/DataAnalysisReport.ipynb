```python
# Select all numerical columns except specified ones
exclude_cols = ["EmployeeCount", "EmployeeNumber", "StandardHours"]
numerical_df = df.select_dtypes(include=['number']).drop(columns=exclude_cols, errors='ignore')
# Step 2: Calculate correlation matrix
corr_matrix = numerical_df.corr()
```

```python
threshold = 0.3 

# Mask values: Remove self-correlation coloring and highlight only significant correlations
masked_corr = np.where((abs(corr_matrix) >= threshold) & (corr_matrix != 1), corr_matrix, np.nan)

# Create heatmap
fig = ff.create_annotated_heatmap(
    z=masked_corr,
    x=corr_matrix.columns.tolist(),
    y=corr_matrix.index.tolist(),
    colorscale="Viridis",
    annotation_text=corr_matrix.round(2).values,  # Keep annotations unchanged
    showscale=True
)

# Update layout to enforce square tiles
fig.update_layout(
    autosize=False,
    width=1200,  # Adjust width
    height=1200,  # Adjust height to match width for square proportions
    xaxis=dict(scaleanchor="y"),  # Keep aspect ratio 1:1
    yaxis=dict(scaleanchor="x")
)

# Show the figure
fig.show()
```

```python
sns.countplot(x=df['Attrition'])
plt.title("Attrition Count")
plt.show()
```

```python
for col in ["Age", "DistanceFromHome", "MonthlyIncome", "YearsAtCompany", "TotalWorkingYears"]:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
cat_cols = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "Over18", "OverTime"]
for col in cat_cols:
    contingency_table = pd.crosstab(df[col], df['Attrition'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"{col}: p-value = {p:.4f}")
```
